Getting book links using BeautifulSoup

```python
def get_book_link(page_no):
    url = 'https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_'+str(page_no)+'?ie=UTF8&pg='+str(page_no)

    r = requests.get(url)

    soup = BeautifulSoup(r.content, 'html.parser')
    
    complete = []
    for d in soup.find_all('div',attrs={'class':"zg-grid-general-faceout"}):
        book_name = d.find('div', attrs={'class':"_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y"})
        link = d.find('a', attrs={'class':"a-link-normal"})

        sub = []
        if book_name is not None:
            sub.append(book_name.text)
        else:
            sub.append('Book name not available')
        if link is not None:
            sub.append('https://www.amazon.in'+link.get('href'))
        else:
            sub.append('Link not available')
        complete.append(sub)
    return complete
```

```python
data = []
no_of_pages = 2
for page in range(1,no_of_pages+1):
    data.append(get_book_link(page))

data
```

Getting book details using Selenium

```python
def clean_book_output(output):
    words = []
    word = ''
    for letter in output:
        if letter != ' ' and letter != ':':
            if letter != '\n':
                word += letter
            else:
                words.append(word)
                words.append(letter)
                word = ''
        else:
            if len(word) != 0:
                words.append(word)
                word = ''

    publisher = []
    language = []
    paperback = []
    isbn10 = []
    isbn13 = []
    weight = []
    dimensions = []

    index = 0

    if 'Publisher' not in words:
        publisher.append('Nan')
    else:
        publisher_index = words.index('Publisher') + 1
        while index != publisher_index:
            index += 1
        while words[index] != '\n':
            publisher.append(words[index])
            index += 1


    if 'Language' not in words:
        language.append('Nan')
    else:
        language_index = words.index('Language') + 1
        while index != language_index:
            index += 1
        while words[index] != '\n':
            language.append(words[index])
            index += 1

    if 'Paperback' not in words:
        paperback.append('Nan')
    else:
        paperback_index = words.index('Paperback') + 1
        while index != paperback_index:
            index += 1
        while words[index] != '\n':
            paperback.append(words[index])
            index += 1

    if 'ISBN-10' not in words:
        isbn10.append('Nan')
    else:
        isbn10_index = words.index('ISBN-10') + 1
        while index != isbn10_index:
            index += 1
        while words[index] != '\n':
            isbn10.append(words[index])
            index += 1

    if 'ISBN-13' not in words:
        isbn13.append('Nan')
    else:
        isbn13_index = words.index('ISBN-13') + 1
        while index != isbn13_index:
            index += 1
        while words[index] != '\n':
            isbn13.append(words[index])
            index += 1

    if 'Item' not in words and 'Weight' not in words:
        weight.append('Nan')
    else:
        weight_index = words.index('Weight') + 1
        while index != weight_index:
            index += 1
        while words[index] != '\n':
            weight.append(words[index])
            index += 1

    if 'Dimensions' not in words:
        dimensions.append('Nan')
    else:
        dimensions_index = words.index('Dimensions') + 1
        while index != dimensions_index:
            index += 1
        while words[index] != '\n':
            dimensions.append(words[index])
            index += 1

    final = []
    final.append(' '.join(publisher))
    final.append(' '.join(language))
    final.append(' '.join(paperback))
    final.append(' '.join(isbn10))
    final.append(' '.join(isbn13))
    final.append(' '.join(weight))
    final.append(' '.join(dimensions))

    return final
```

```python
def get_book_details_sl(link):
    driver.get(link)

    details = []
    elemetn = driver.find_element(By.ID, "detailBullets_feature_div")
    output = clean_book_output(elemetn.text)
    for i in output:
        details.append(i)
    
    reviews = driver.find_elements(By.XPATH, "//div[@class='a-expander-content reviewText review-text-content a-expander-partial-collapse-content']/span")
    count = 0
    for r in reviews:
        details.append(r.text)
        count += 1
        if count == 3:
            break
    return details
```

Getting movie details using Beautiful Soup

```python
def get_movie_detail_bs(movie_no):
    url_bs = 'https://www.imdb.com/search/title/?release_date=2020-01-01,2020-12-31&sort=num_votes,desc&start='+ str(movie_no) +'&ref_=adv_nxt'
    r = requests.get(url_bs)
    soup = BeautifulSoup(r.content, 'html.parser')
    details_bs = []
    for d in soup.find_all('div', attrs={'class':'lister-item mode-advanced'}):
        name_bs = d.find('h3', attrs={'class':'lister-item-header'})
        rating_bs = d.find('div', attrs={'class':'inline-block ratings-imdb-rating'})
        desc_bs = d.find_all('p', attrs={'class':'text-muted'})
        votes_bs = d.find('span', attrs={'name':'nv'})
        directors_stars_bs = d.find('p', attrs={'class':''})
        metascore_bs = d.find('div', attrs={'class':'inline-block ratings-metascore'})
        genre_bs = d.find('span', attrs={'class':'genre'})
        total_duration_bs = d.find('span', attrs={'class':'runtime'})

        sub_lst_bs = []
        if name_bs is not None:
            sub_lst_bs.append(name_bs.a.text)
        else:
            sub_lst_bs.append('Nan')
        if rating_bs is not None:
            sub_lst_bs.append(rating_bs.strong.text)
        else:
            sub_lst_bs.append('Nan')
        if desc_bs is not None:
            sub_lst_bs.append(desc_bs[1].text.strip())
        else:
            sub_lst_bs.append('Nan')
        if votes_bs is not None:
            sub_lst_bs.append(votes_bs.text)
        else:
            sub_lst_bs.append('Nan')
    
        directors_bs = []
        stars_bs = []
        words_bs = []
        if directors_stars_bs is not None:
            output = directors_stars_bs.text.strip()
            word = ''
            for letter in output:
                if letter != " " and letter != '\n' and letter != '|':
                    word += letter
                else:
                    if len(word) != 0:
                        words_bs.append(word)
                        word = ''
            flag = True
            for word in words_bs:
                if word != 'Stars:' and flag:
                    if word != 'Director:':
                        directors_bs.append(word)
                else:
                    if flag:
                        if len(directors_bs) == 0:
                            sub_lst_bs.append('Nan')
                        else:
                            sub_lst_bs.append(' '.join(directors_bs))
                        flag = False
                        continue
                    stars_bs.append(word)
            if len(stars_bs) == 0:
                sub_lst_bs.append('Nan')
            else:
                sub_lst_bs.append(' '.join(stars_bs))

        else:
            sub_lst_bs.append('Nan')
            sub_lst_bs.append('Nan')

        if metascore_bs is not None:
            sub_lst_bs.append(metascore_bs.span.text.strip())
        else:
            sub_lst_bs.append('Nan')
        if genre_bs is not None:
            sub_lst_bs.append(genre_bs.text.strip())
        else:
            sub_lst_bs.append('Nan')
        if total_duration_bs is not None:
            sub_lst_bs.append(total_duration_bs.text)
        else:
            sub_lst_bs.append('Nan')
            
        details_bs.append(sub_lst_bs)
    return details_bs
    
```

```python
details_bs = []
page_bs = 1
count_bs = 0
while(count_bs < 1000):
    details_bs.append(get_movie_detail_bs(count_bs + 1))
    count_bs += len(details_bs[page_bs - 1])
    page_bs += 1

details_bs
```

```python
def flatten_bs(list):
    final_details = []
    for page in list:
        for movie in page:
            final_details.append(movie)
    return final_details
```

```python
df_movies_bs = pd.DataFrame(flatten_bs(details_bs), columns=['Movie Name', 'Ratings', 'Desctiption', 'Votes', 'Director(s)', 'Stars', 'Metascore', 'Movie Genre', 'Total Duration'])
df_movies_bs
```

```python
def get_movie_detail_sl(movie_no):
    url_sl = 'https://www.imdb.com/search/title/?release_date=2020-01-01,2020-12-31&sort=num_votes,desc&start='+ str(movie_no) +'&ref_=adv_nxt'
    
    driver.get(url_sl)

    details_sl = []
    for d in driver.find_elements(By.XPATH, "//div[@class='lister-item mode-advanced']"):
        sub_list_sl = []
        try:
            title_sl = d.find_element(By.XPATH, "div[@class='lister-item-content']/h3[@class='lister-item-header']/a")
            sub_list_sl.append(title_sl.text)
        except NoSuchElementException:
            sub_list_sl.append('Nan')

        try:
            rating_sl = d.find_element(By.XPATH, "div[@class='lister-item-content']/div[@class='ratings-bar']/div[@class='inline-block ratings-imdb-rating']/strong")
            sub_list_sl.append(rating_sl.text)
        except NoSuchElementException:
            sub_list_sl.append('Nan')

        try:
            desc_sl = d.find_element(By.XPATH, "div[@class='lister-item-content']/p[@class='text-muted']")
            sub_list_sl.append(desc_sl.text.strip())
        except NoSuchElementException:
            sub_list_sl.append('Nan')

        try:
            votes_sl = d.find_element(By.XPATH, "div[@class='lister-item-content']/p[@class='sort-num_votes-visible']/span[@name='nv']")
            sub_list_sl.append(votes_sl.text)
        except NoSuchElementException:
            sub_list_sl.append('Nan')
        
        directors_sl = []
        stars_sl = []
        words_sl = []
        try:
            directors_stars_sl = d.find_element(By.XPATH, "div[@class='lister-item-content']/p[@class='']")
            output = directors_stars_sl.text.strip()
            word = ''
            for letter in output:
                if letter != " " and letter != '\n' and letter != '|':
                    word += letter
                else:
                    if len(word) != 0:
                        words_sl.append(word)
                        word = ''
            flag = True
            for word in words_sl:
                if word != 'Stars:' and flag:
                    if word != 'Director:':
                        directors_sl.append(word)
                else:
                    if flag:
                        if len(directors_sl) == 0:
                            sub_list_sl.append('Nan')
                        else:
                            sub_list_sl.append(' '.join(directors_sl))
                        flag = False
                        continue
                    stars_sl.append(word)
            if len(stars_sl) == 0:
                sub_list_sl.append('Nan')
            else:
                sub_list_sl.append(' '.join(stars_sl)) 
            
        except NoSuchElementException:
            sub_list_sl.append('Nan')
            sub_list_sl.append('Nan')

        try:
            metascore_sl = d.find_element(By.XPATH, "div[@class='lister-item-content']/div[@class='ratings-bar']/div[@class='inline-block ratings-metascore']/span")
            sub_list_sl.append(metascore_sl.text.strip())
        except NoSuchElementException:
            sub_list_sl.append('Nan')
        
        try:
            genre_sl = d.find_element(By.XPATH, "div[@class='lister-item-content']/p[@class='text-muted ']/span[@class='genre']")
            sub_list_sl.append(genre_sl.text.strip())
        except NoSuchElementException:
            sub_list_sl.append('Nan')

        try:
            total_duration_sl = d.find_element(By.XPATH, "div[@class='lister-item-content']/p[@class='text-muted ']/span[@class='runtime']")
            sub_list_sl.append(total_duration_sl.text)
        except NoSuchElementException:
            sub_list_sl.append('Nan')
        
        details_sl.append(sub_list_sl)
    return details_sl
```
